```python
from pathlib import Path
import pandas as pd
import numpy as np

# Project paths
ROOT_DIR = Path.cwd().parent
DATA_DIR = ROOT_DIR / "data"
RAW_DIR = DATA_DIR / "raw"

print("Root directory:", ROOT_DIR)
print("Data directory:", DATA_DIR)
print("Raw data directory:", RAW_DIR)
```

## Raw Dataset Inventory

The table above shows the raw datasets available in the project.

The project contains:

- World Bank Albania dataset
- World Bank population estimates dataset
- Eurostat migration flow datasets
- Eurostat residence permit datasets
- Official reports from INSTAT and the World Bank

These datasets will be used to analyze Albania's population decline, migration pressure, economic indicators, unemployment, residence permits, and demographic risk.

```python
# List all raw data files

raw_files = []

for file in RAW_DIR.rglob("*"):
    if file.is_file():
        raw_files.append({
            "file_name": file.name,
            "folder": file.parent.name,
            "full_path": str(file.relative_to(ROOT_DIR)),
            "size_mb": round(file.stat().st_size / (1024 * 1024), 2)
        })

raw_files_df = pd.DataFrame(raw_files)
raw_files_df
```
